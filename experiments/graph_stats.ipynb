```python
import numpy as np
import scipy as sp

from tqdm import tqdm

from utils import *
```

```python
names = ["football", "primary1", "primary2", "eu-core", "eurosis", "cora_small", "airport", "blogcatalog", "cora", "as"]
```

```python
for name in names:
    adj, lab = load_graph(name, force_sparse=True)

    n = adj.shape[0]
    m = len(adj.data) // 2
    c = lab.shape[0]
    nout = np.sum(lab.count_nonzero(axis=0) == 0)
    n_labels = np.asarray(lab.sum(axis=0))
    eta = np.mean(n_labels[n_labels > 0])
    in_com = 0
    labcsc = lab.tocsc()
    for i,j in zip(*adj.nonzero()):
        i_coms = labcsc[:, i].col
        j_coms = labcsc[:, j].col
        if len(np.intersect1d(i_coms, j_coms)) > 0:
            in_com += 1
    xi = 1 - in_com / len(adj.data)

    print(f"{name.capitalize()} & & {n} & {m} & {c} & {nout} & {eta:.2f} & {xi:.2f} \\\\ ")
```
